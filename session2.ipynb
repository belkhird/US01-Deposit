```python
iso_codes={}
```

iso_codes est une variables et a chaque fois qu'on va executer le bloc ca va mettre la variable à 0 chose que nous voulons pas forcement donc on va slit cell pour éviter de la remettre à 0 à chaque exécution 

```python
for index, name in myfile['Country (according to CN Gov):'].items():
    if name in iso_codes:
        continue
        
    cleaned_name=name.split('(')[0].strip()
    if cleaned_name == 'Thailiand':
        cleaned_name='Thailand'
        
    if cleaned_name == 'The United Arab Emirates':
        cleaned_name='United Arab Emirates'
        
    url= 'https://restcountries.eu/rest/v2/name/'+cleaned_name
    r= requests.get(url)
    if r.status_code==200:
        data= r.json()
        country_code=data[0]['alpha3Code']
        iso_codes[name]=country_code
    else:
        print('[ERROR] WE CANNOT GET GET THE DATA, REASON IS', r.status_code, r.text)
        print(name)
    
            
```

```python
valeurs={}
```

```python
    import time
```

```python
for index, name in myfile['isocode'].items():
    print(name)
    
    if name in valeurs:
        continue
         
    time.sleep(5)
    
    if name == 'IOT':
        url= 'http://api.worldbank.org/v2/country/' + 'IND' + '/indicator/GC.TAX.IMPT.ZS?format=json' 
    else:
        url= 'http://api.worldbank.org/v2/country/' + name + '/indicator/GC.TAX.IMPT.ZS?format=json'
    r= requests.get(url)
    if r.status_code==200:
        data= r.json()
        compteur=0
        
        if name == 'BRN' or name == 'LAO'or name == 'MNE'or name == 'OMN' or name == 'PAK' or name == 'QAT' or name == 'SYR' or name == 'TKM' or name == 'VNM' or name == 'YEM':
            valeurs[name]="No value"
        else:
            valeurs[name]=data[1][compteur]['value']
        
        while valeurs[name] is None:
            compteur += 1
            valeurs[name]=data[1][compteur]['value']
            
    else:
        print('[ERROR] WE CANNOT GET GET THE DATA, REASON IS', r.status_code, r.text)
        print(name)
```
